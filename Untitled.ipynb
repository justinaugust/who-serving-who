```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import PolynomialFeatures
```

```python
%store -r all_years_data
%store -r subgroups
%store -r tests
%store -r meta
```

```python
all_years_data.head()
```

```python
[all_years_data.drop(columns=column, inplace=True) for column in all_years_data.columns if '_tested' in column]
all_years_data.drop(columns=['school_name','lat','lng','year'], inplace=True)
```

```python
all_years_data.set_index("school_id", inplace=True)
all_years_data.fillna(0, inplace=True)
```

```python
test_1_columns = [column for column in all_years_data.columns if '_test_1' in column]
```

```python
test_1_to_sparse = all_years_data.loc[:,test_1_columns].copy()
test_1_to_sparse.head()
```

```python
poly = PolynomialFeatures(interaction_only=True)
test_1_to_sparse_poly = poly.fit_transform(test_1_to_sparse)
test_1_to_sparse_poly
```

```python
test_1_sparse = sparse.csr_matrix(test_1_to_sparse.fillna(0))

test_1_recommender = pairwise_distances(test_1_sparse, metric='cosine')

test_1_recommender_df = pd.DataFrame(test_1_recommender, index=test_1_to_sparse.index, columns=test_1_to_sparse.index)

test_1_recommender_df
```
